# **Boards**

```python
with open(path+'boards.txt', 'r') as file:
    input = file.read().splitlines()

unwanted_sign_in_boards = []
for i in range((len(input))):
  unwanted_sign_in_boards.append(input[i].split(' '))

without_unwanted = []
for k in range(len(unwanted_sign_in_boards)):
  for i in range(len(unwanted_sign_in_boards[k])):
    if unwanted_sign_in_boards[k][i] != '':
      #print(a[k][i])
      without_unwanted.append(unwanted_sign_in_boards[k][i])
    else:
      pass
  #print(a[4][i])

boards = np.array(without_unwanted, dtype=int)

boards = boards.reshape([100,5,5])

boards
```

# **Part 1**
**Finding the first winner board**

```python
boards_all = boards.copy()
numbers_all = numbers.copy()
buffer = np.zeros([100,5,5])

breaking = 0
round = 0

Last_Number = 0
Winner_Board = 0


for h in numbers_all.iloc[0, :]:

  for i in range(len(boards_all)):

    for j in range(len(boards_all[0])):
      count_row = 0
      count_column = 0

      for k in range(len(boards_all[0][0])):
        if boards_all[i][j][k] == h:
          buffer[i][j][k] += 1
        
      count_row += buffer[i][j].sum()
      count_column += np.transpose(buffer, (0, 2, 1))[i][j].sum()
      if count_row == 5 or count_column == 5:
        
        breaking = 1
        Last_Number = h
        Winner_Board = i

        print('Round =', round)
        print('Last number =', h)
        print('Winner board =', i)
        if count_row == 5: # and count_row == 5:
          print('By row')
        elif count_column == 5:
          print('By column')
        break

  if breaking == 1:
    break
  
  round +=1
  
```

```python
unmarked_list = []
for i in range(len(buffer[Winner_Board])):
  for j in range(len(buffer[Winner_Board][0])):
    if buffer[Winner_Board][i][j] == 0:
      unmarked_list.append(boards[Winner_Board][i][j])

result = (np.sum(unmarked_list))*Last_Number

print('Winner Board : \n', boards[Winner_Board])
print('Unmarked list of the winner board :', unmarked_list)
print('Result =', result)
```

# **Part 2**
**Finding the last winner board**

```python
boards_all = boards.copy()
numbers_all = numbers.copy()
buffer = np.zeros([100,5,5])

breaking = 0
round = 0

Last_Number = 0
Last_Winner_Board = 0

All_winners = []

for h in numbers_all.iloc[0, :]:

  for i in range(len(boards_all)):

    for j in range(len(boards_all[0])):
      count_row = 0
      count_column = 0

      for k in range(len(boards_all[0][0])):
        if boards_all[i][j][k] == h:
          buffer[i][j][k] += 1
        
      count_row += buffer[i][j].sum()
      count_column += np.transpose(buffer, (0, 2, 1))[i][j].sum()
      if count_row == 5 or count_column == 5:
        
        All_winners.append(i)
    
    if len(set(All_winners)) == 100:
      breaking = 1
      Last_Number = h
      Last_Winner_Board = i

      print('Round =', round)
      print('Last number =', h)
      print('The last winner board =', i)
        #if count_row == 5: # and count_row == 5:
        #  print('By row')
        #elif count_column == 5:
        #  print('By column')
      break

  if breaking == 1:
    break
  
  round +=1
```

```python
unmarked_list = []
for i in range(len(buffer[Last_Winner_Board])):
  for j in range(len(buffer[Last_Winner_Board][0])):
    if buffer[Last_Winner_Board][i][j] == 0:
      unmarked_list.append(boards[Last_Winner_Board][i][j])

result = (np.sum(unmarked_list))*Last_Number

print('Winner Board : \n', boards[Last_Winner_Board])
print('Unmarked list of the winner board :', unmarked_list)
print('Result =', result)
```
