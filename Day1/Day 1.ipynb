```python
# Import data
path = '/content/drive/My Drive/Colab Notebooks/Udemy_Complete Machine Learning and Data Science/AoC/Day 1/' # create path to call file(s) in my directory 
inputs = pd.read_csv(path+'inputs.txt', header = None) # create a dataframe of the input data
inputs
```

```python
# Counting the longer measurements
count = 0 # create a count variable at 0
for i in range(len(inputs)): # check all values in the entire inputs data frame
  if i+1 == 2000: # if the last position of the datapoints reach 2000 which exceed the range of the data, then break the loop
    break
  elif inputs.loc[i+1,0]>inputs.loc[i,0]: # if the new meassurement is larger than the previous one, counting 1 in the count variable
    count +=1
  elif inputs.loc[i+1,0]<=inputs.loc[i,0]: # if the new meassurement is smaller or equal than the previous one, counting 1 in the count variable
    count +=0

count # represent the answer
```
