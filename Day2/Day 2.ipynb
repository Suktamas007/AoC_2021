# **Import and Process the data**

```python
# Import data
path = '/content/drive/My Drive/Colab Notebooks/Udemy_Complete Machine Learning and Data Science/AoC/Day 2/' # create path to call file(s) in my directory 
inputs = pd.read_csv(path+'input.txt', header = None) # create a dataframe of the input data
inputs
```

```python
# Split data into two collumns called Directions and Quantity
inputs[['Directions','Quantity']]= inputs[0].str.split(" ", n = 2, expand = True)
inputs = inputs.drop(columns=0)
inputs.head(10)
```

```python
inputs.dtypes # Show the data types of each column
```

```python
# Change type of a Quantity column from object to int type in order to be able to calculate them
inputs['Quantity'] = inputs['Quantity'].astype(int)
```

```python
inputs.dtypes
```

# **Part 1**

```python
forward = 0
depth = 0

for i in range(len(inputs)): 
  if i == 1000: 
    break
  elif inputs.loc[i][0] == 'forward':
    forward += inputs.loc[i][1]
  elif inputs.loc[i][0] == 'up':
    depth -= inputs.loc[i][1]
  elif inputs.loc[i][0] == 'down':
    depth += inputs.loc[i][1]

print('Total forward =', forward)
print('Total depth =', depth)
result = forward*depth
print('Result of the total forward multiplied by the total depth =', result)
```

```python
forward = 0
aim = 0
depth = 0

for i in range(len(inputs)): #range(6)
  if i == 1000: 
    break
  elif inputs.loc[i][0] == 'forward':
    forward += inputs.loc[i][1]
    depth += aim*inputs.loc[i][1]
  elif inputs.loc[i][0] == 'up':
    aim -= inputs.loc[i][1]
  elif inputs.loc[i][0] == 'down':
    aim += inputs.loc[i][1]

print('Total forward =', forward)
print('Total depth =', depth)
result = forward*depth
print('Result of the total forward multiplied by the total depth =', result)
```
